```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
sns.set()
```

```python
stock = pd.read_csv('Tesla.csv - Tesla.csv.csv')
stock.head()
stock.set_index('Date', inplace=True)
stock
```

```python
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
fig.suptitle('box plot for all fields')
sns.boxplot(ax=axes[0, 0], data=stock['Open'])
sns.boxplot(ax=axes[0, 1], data=stock['High'])
sns.boxplot(ax=axes[0, 2], data=stock['Low'])
sns.boxplot(ax=axes[1, 0], data=stock['Close'])
sns.boxplot(ax=axes[1, 1], data=stock['Volume'])
sns.boxplot(ax=axes[1, 2], data=stock['Adj Close'])
```

```python
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
fig.suptitle('box plot for all fields')
sns.lineplot(ax=axes[0, 0], data=stock['Open'])
sns.lineplot(ax=axes[0, 1], data=stock['High'])
sns.lineplot(ax=axes[0, 2], data=stock['Low'])
sns.lineplot(ax=axes[1, 0], data=stock['Close'])
sns.lineplot(ax=axes[1, 1], data=stock['Volume'])
sns.lineplot(ax=axes[1, 2], data=stock['Adj Close'])
```

```python
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
fig.suptitle('scatter plot for all fields')
sns.scatterplot(ax=axes[0, 0], data=stock['Open'])
sns.scatterplot(ax=axes[0, 1], data=stock['High'])
sns.scatterplot(ax=axes[0, 2], data=stock['Low'])
sns.scatterplot(ax=axes[1, 0], data=stock['Close'])
sns.scatterplot(ax=axes[1, 1], data=stock['Volume'])
sns.scatterplot(ax=axes[1, 2], data=stock['Adj Close'])
```

```python
r = stock['Volume'].quantile(0.9)
stock_1 = stock[stock['Volume'] < r]
stock_1
```

```python
r = stock_1['Open'].quantile(0.9)
stock_2 = stock_1[stock_1['Open'] < r]
stock_2
```

```python
r = stock_2['Close'].quantile(0.9)
stock_3 = stock_2[stock_2['Close'] < r]
stock_3
```

```python
r = stock_3['High'].quantile(0.9)
stock_4= stock_3[stock_2['High'] < r]
stock_4
```

```python

r = stock_4['Low'].quantile(0.9)
stock_5 = stock_4[stock_4['Low'] < r]
stock_5
```

```python
g, (x1,x3,x4) = plt.subplots(1,3, sharey=True, figsize=(15,3))
x1.scatter(stock_5['Open'],stock_5['Close'])
x1.set_title('Open and Close')
x3.scatter(stock_5['High'],stock_5['Close'])
x3.set_title('High and Close')
x4.scatter(stock_5['Low'],stock_5['Close'])
x4.set_title('Low and Close')
plt.show()
```

```python
x = stock_5[['Open','High','Low','Volume']]
y = stock_5['Close']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size= 0.2)
```

```python
reg = LinearRegression()
reg.fit(x_train,y_train)
```

```python
pred_1=reg.predict(x_train)
pred_1
```

```python
sns.scatterplot( data=prediction1)
```
